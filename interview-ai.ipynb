# Write a function analyze_value(value) that takes ANY Python value
# and returns a formatted string with:
# - The value itself
# - Its type
# - Its truthiness (True/False)
# - Its length (if applicable, "N/A" otherwise)
# Expected usage:
# analyze_value(42) → "Value: 42 | Type: int | Truthy: True | Length: N/A"
# analyze_value("") → "Value: | Type: str | Truthy: False | Length: 0"
# analyze_value([1,2,3]) → "Value: [1, 2, 3] | Type: list | Truthy: True | Length: 3"

````python
def analyze_value(value):
    """
    Analyzes a Python value and returns a formatted summary string.

    Parameters:
        value (any): The Python value to analyze.

    Returns:
        str: Formatted analysis string.
    """

    # Determine type
    value_type = type(value).__name__

    # Determine truthiness
    is_truthy = bool(value)

    # Determine length safely
    try:
        value_length = len(value)
    except TypeError:
        value_length = "N/A"

    return (
        f"Value: {value} | "
        f"Type: {value_type} | "
        f"Truthy: {is_truthy} | "
        f"Length: {value_length}"
    )
analyze_value(9)
````

````python
Below is a **complete Python type conversion matrix** for:

```
int, float, str, bool, list, tuple
```

All conversions are shown explicitly.
Invalid conversions include the **exact error type**.
Edge cases and dangerous behaviors are highlighted.

---

# 🔷 1. INT → Other Types

Assume:

```python
x = 5
zero = 0
```

| Conversion     | Syntax     | Result      | Why                   |
| -------------- | ---------- | ----------- | --------------------- |
| int → float    | `float(5)` | `5.0`       | Numeric widening      |
| int → str      | `str(5)`   | `"5"`       | String representation |
| int → bool     | `bool(5)`  | `True`      | Non-zero is truthy    |
| int → bool (0) | `bool(0)`  | `False`     | 0 is falsy            |
| int → list     | `list(5)`  | ❌ TypeError | int not iterable      |
| int → tuple    | `tuple(5)` | ❌ TypeError | int not iterable      |

---

# 🔷 2. FLOAT → Other Types

Assume:

```python
x = 3.9
zero = 0.0
```

| Conversion         | Syntax       | Result      | Why                   |
| ------------------ | ------------ | ----------- | --------------------- |
| float → int        | `int(3.9)`   | `3`         | Truncation toward 0   |
| float → str        | `str(3.9)`   | `"3.9"`     | String representation |
| float → bool       | `bool(3.9)`  | `True`      | Non-zero is truthy    |
| float → bool (0.0) | `bool(0.0)`  | `False`     | 0.0 is falsy          |
| float → list       | `list(3.9)`  | ❌ TypeError | Not iterable          |
| float → tuple      | `tuple(3.9)` | ❌ TypeError | Not iterable          |

⚠️ Important:

```python
int(-3.9)  # -3 (not -4)
```

Truncates toward zero, not floor.

---

# 🔷 3. STR → Other Types

Assume:

```python
s = "10"
empty = ""
space = "  10  "
bad = "abc"
```

| Conversion             | Syntax          | Result          | Why                        |
| ---------------------- | --------------- | --------------- | -------------------------- |
| str → int              | `int("10")`     | `10`            | Valid numeric string       |
| str → int (whitespace) | `int("  10  ")` | `10`            | Whitespace ignored         |
| str → int ("abc")      | `int("abc")`    | ❌ ValueError    | Invalid literal            |
| str → float            | `float("3.14")` | `3.14`          | Valid float string         |
| str → float ("abc")    | `float("abc")`  | ❌ ValueError    | Invalid literal            |
| str → bool             | `bool("False")` | `True`          | Non-empty string is truthy |
| str → bool (empty)     | `bool("")`      | `False`         | Empty is falsy             |
| str → list             | `list("abc")`   | `['a','b','c']` | Iterable                   |
| str → tuple            | `tuple("abc")`  | `('a','b','c')` | Iterable                   |

🔥 Dangerous Behavior:

```python
bool("False")  # True
bool("0")      # True
```

Because only empty string is falsy.

---

# 🔷 4. BOOL → Other Types

Recall:

```python
True == 1
False == 0
```

| Conversion   | Syntax         | Result      | Why                   |
| ------------ | -------------- | ----------- | --------------------- |
| bool → int   | `int(True)`    | `1`         | bool subclass of int  |
| bool → int   | `int(False)`   | `0`         |                       |
| bool → float | `float(True)`  | `1.0`       | Numeric promotion     |
| bool → str   | `str(True)`    | `"True"`    | String representation |
| bool → list  | `list(True)`   | ❌ TypeError | Not iterable          |
| bool → tuple | `tuple(False)` | ❌ TypeError | Not iterable          |

🔎 Important:

```python
isinstance(True, int)  # True
```

`bool` is a subclass of `int`.

---

# 🔷 5. LIST → Other Types

Assume:

```python
lst = [1, 2]
single = [10]
empty = []
nested = [[1,2], [3,4]]
```

| Conversion          | Syntax         | Result      | Why                 |
| ------------------- | -------------- | ----------- | ------------------- |
| list → int          | `int([1,2])`   | ❌ TypeError | Invalid type        |
| list → float        | `float([1,2])` | ❌ TypeError | Invalid type        |
| list → str          | `str([1,2])`   | `"[1, 2]"`  | String repr         |
| list → bool         | `bool([1,2])`  | `True`      | Non-empty truthy    |
| list → bool (empty) | `bool([])`     | `False`     | Empty falsy         |
| list → tuple        | `tuple([1,2])` | `(1,2)`     | Iterable conversion |

⚠ Multiple-element numeric conversion:

```python
int([10])   # ❌ TypeError
```

Even single element list fails.

---

# 🔷 6. TUPLE → Other Types

Assume:

```python
t = (1,2)
single = (10,)
empty = ()
```

| Conversion           | Syntax         | Result      | Why                 |
| -------------------- | -------------- | ----------- | ------------------- |
| tuple → int          | `int((1,2))`   | ❌ TypeError | Invalid type        |
| tuple → float        | `float((1,2))` | ❌ TypeError | Invalid type        |
| tuple → str          | `str((1,2))`   | `"(1, 2)"`  | String repr         |
| tuple → bool         | `bool((1,2))`  | `True`      | Non-empty truthy    |
| tuple → bool (empty) | `bool(())`     | `False`     | Empty falsy         |
| tuple → list         | `list((1,2))`  | `[1,2]`     | Iterable conversion |

---

# 🔥 Nested Structures

```python
list(((1,2), (3,4)))
# [(1,2), (3,4)]
```

Conversion preserves nesting level.

```python
int((10,))
# ❌ TypeError
```

Even single-element tuple cannot be cast directly.

---

# 🔥 High-Risk / Surprising Behaviors

| Expression         | Result    | Why Dangerous            |
| ------------------ | --------- | ------------------------ |
| `bool("False")`    | True      | Non-empty string         |
| `bool("0")`        | True      | Same reason              |
| `int(True)`        | 1         | bool subclass of int     |
| `0.1 + 0.2 == 0.3` | False     | Floating point precision |
| `int(3.99)`        | 3         | Truncation, not rounding |
| `list("10")`       | ['1','0'] | Character iteration      |

---

# 🔎 Why Some Conversions Fail

### Rule 1 — Numeric constructors

`int()` and `float()` accept:

* Numbers
* Numeric strings
* Objects implementing `__int__` / `__float__`

Lists/tuples do not qualify.

---

### Rule 2 — list() and tuple()

They require:

* Iterable input

`int`, `float`, `bool` are not iterable.

---

### Rule 3 — bool()

`bool(x)` checks truthiness:

Falsy:

* 0
* 0.0
* ""
* []
* ()
* None
* {}

Everything else is True.

---

# 🧠 Conceptual Summary

Python type conversion falls into 4 categories:

1. Numeric promotion/truncation
2. String parsing
3. Iterable unpacking
4. Truth-value testing

Failures occur when:

* Object is not iterable
* String cannot be parsed
* No conversion protocol exists

````
